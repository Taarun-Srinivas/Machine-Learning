```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```
path = '/gdrive/My Drive/ML:Pilot/Assignments/Data/Fish.csv'
raw_data = pd.read_csv(path)
raw_data.head()
```

```
raw_data.shape
```

```
y = raw_data.iloc[:,1:2]
y
```

```
x = raw_data.drop(columns = ['Weight'])
x
```

```
from sklearn.preprocessing import LabelEncoder 

labelencoder = LabelEncoder()

x['Species'] = labelencoder.fit_transform(x['Species'])
x.head()
```

```
x.shape, y.shape
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x)
y_train = scaler.fit_transform(y)
```

```
x_train.shape, y_train.shape
```

```
from sklearn.linear_model import LinearRegression 
lreg = LinearRegression()
```

```
lreg.fit(x_train,y_train)
```

```
train_pred = lreg.predict(x_train)
```

```
from sklearn.metrics import r2_score
print(r2_score(y_train, train_pred))
```

```
import math 
erro_r = [np.power((b-a),2) for (a,b) in zip (train_pred, y_train)]
error0 = np.sum(erro_r)
error = math.sqrt(error0)
error = (error/ len(y))*100

print("training_error % = {}".format(error))
accuracy = 100 - error
print("training_accuracy % = {}".format(accuracy))
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_test = scaler.fit_transform(x)
y_test = scaler.fit_transform(y)
```

```
x_test.shape, y_test.shape
```

```
test_pred = lreg.predict(x_test)
```

```
from sklearn.metrics import r2_score
print(r2_score(y_test, test_pred))
```

```
import math 
erro_r = [np.power((b-a),2) for (a,b) in zip (test_pred, y_test)]
error0 = np.sum(erro_r)
error = math.sqrt(error0)
error = (error/ len(y))*100

print("testing error % = {}".format(error))
accuracy = 100 - error
print("testing accuracy % = {}".format(accuracy))
```
