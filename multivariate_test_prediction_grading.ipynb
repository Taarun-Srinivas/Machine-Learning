```
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```
path = '/gdrive/My Drive/ML:Pilot/Assignments/Data/Chennai_house_price_multivariate_train.csv'
```

```
raw_data = pd.read_csv(path)
raw_data.head()
```

```
raw_data.shape
```

```
raw_data.isnull().sum(axis=0)
```

```
raw_data = raw_data.fillna({"N_BEDROOM": raw_data["N_BEDROOM"].mean(), "N_BATHROOM": raw_data["N_BATHROOM"].mean(), "QS_OVERALL": raw_data["QS_OVERALL"].mean()})
```

```
raw_data
```

```
raw_data.isnull().sum(axis = 0)
```

```
y = raw_data['SALES_PRICE'] 
y = np.array(y)
y = y.reshape(-1,1)
y.shape
```

```
x_raw_data = raw_data.drop(columns=['PRT_ID','SALES_PRICE','DATE_SALE','DATE_BUILD'])
x_raw_data.head()
```

```
y.shape
```

```
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
x_raw_data["AREA"] = labelencoder.fit_transform(x_raw_data["AREA"])
x_raw_data["SALE_COND"] = labelencoder.fit_transform(x_raw_data["SALE_COND"])
x_raw_data["PARK_FACIL"] = labelencoder.fit_transform(x_raw_data["PARK_FACIL"])
x_raw_data["BUILDTYPE"] = labelencoder.fit_transform(x_raw_data["BUILDTYPE"])
x_raw_data["UTILITY_AVAIL"] = labelencoder.fit_transform(x_raw_data["UTILITY_AVAIL"])
x_raw_data["STREET"] = labelencoder.fit_transform(x_raw_data["STREET"])
x_raw_data["AREA"] = labelencoder.fit_transform(x_raw_data["AREA"])
x_raw_data["MZZONE"] = labelencoder.fit_transform(x_raw_data["MZZONE"])
```

```
x_raw_data.head()
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scalery = MinMaxScaler()
x = scaler.fit_transform(x_raw_data)
y = scalery.fit_transform(y)
```

```
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, shuffle = False)
```

```
from sklearn.linear_model import LinearRegression 
lreg = LinearRegression()
```

```
lreg.fit(x_train,y_train)
```

```
train_pred = lreg.predict(x_train)
```

```
from sklearn.metrics import r2_score
print(r2_score(y_train, train_pred))
```

```
import math 
erro_r = [np.power((b-a),2) for (a,b) in zip (train_pred, y_train)]
error0 = np.sum(erro_r)
error = math.sqrt(error0)
error = (error/ len(y))*100

print("training_error % = {}".format(error))
accuracy = 100 - error
print("training_accuracy % = {}".format(accuracy))
```

```
test_pred = lreg.predict(x_test)
```

```
from sklearn.metrics import r2_score
print(r2_score(y_test, test_pred))
```

```
import math 
erro_r = [np.power((b-a),2) for (a,b) in zip (test_pred, y_test)]
error0 = np.sum(erro_r)
error = math.sqrt(error0)
error = (error/ len(y))*100

print("testing error % = {}".format(error))
accuracy = 100 - error
print("testing accuracy % = {}".format(accuracy))
```
